2. Find the type of contour that has the maximum productivity?

```python
a=soil[soil['P']==soil["P"].max()]
print(a)
a['Contour']
```

3. What is the minimum productivity of Depression contour?

```python
dep=soil[soil['Contour']=='Depression']

print("Minimum productiviety with depression Contour is ",dep['P'].min())
```

4. Is a scatter plot suitable to display the potassium level of each contour? If yes, use
scatter plot, else use appropriate charts.

```python
soil.plot(x="Contour",y="P", kind="scatter", color="r", fontsize="10")
plt.title("Scatter Plot - Soil")
plt.xlabel("Potassium Level")
plt.ylabel("Contour")
plt.show()
```

5. Use pie chart to display the average density level of every contour.

```python
e=(0.01,0.01,0.01)
Avg=list(soil.groupby('Contour')['Dens'].mean())
plt.pie(Avg,explode=e,labels=['Top','Slope','Depression'])
plt.title('Pie Chart')
plt.legend(title="Soil Contour Density")
plt.legend(loc="upper right")
plt.show()
```
