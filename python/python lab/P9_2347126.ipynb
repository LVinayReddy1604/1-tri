Histogram

```python
male["Experience_Years"].plot(kind="hist", label="Male Salary", color="blue", fontsize="5", density=False, alpha=0.5)
female['Experience_Years'].plot(kind="hist", label="Female Salary", color="pink", fontsize="5", density=False, alpha=0.5)
plt.title("Scatter Plot - Histogram")
plt.xlabel("Salary")
plt.show()
```

Scatter-Plot

```python
male.plot(x="Experience_Years",y="Salary", kind="scatter", label="Male", color="b", fontsize="10")
plt.title("Scatter Plot - Male")
plt.xlabel("Experience Years")
plt.ylabel("Salary")
plt.show()
```

```python
female.plot(x="Experience_Years",y="Salary", kind="scatter", label="Female", color="r", fontsize="10")
plt.title("Scatter Plot - Female")
plt.xlabel("Experience Years")
plt.ylabel("Salary")
plt.show()
```

```python
male['Experience_Years'].plot(kind='line',color="blue",label="Male", linewidth='1.5',ls="-")
female['Experience_Years'].plot(kind='line',color="pink",label="Female",linewidth='1.5',ls=":")
plt.title("Line Plot - Experience Years")
plt.xlabel("ID")
plt.ylabel("Years")
plt.show()
```

Bar-Plot

```python
Type=list(data['Gender'].unique())
count=list(data['Gender'].value_counts())
Avg=list(data.groupby('Gender')["Salary"].mean())
plt.bar(Type, Avg, color=['Maroon','yellow','blue'],width=0.4)
plt.title('Bar - Plot')
plt.xlabel('Gender')
plt.ylabel('Average Salary')
plt.show()
```

Pie-Chart

```python
e=(0,0.01)
Avg=list(data.groupby('Gender')['Salary'].mean())
plt.pie(Avg,explode=e,labels=['Male','Female'])
plt.title('Pie Chart')
plt.legend(title="Salaries")
plt.legend(loc="upper right")
plt.show()
```

```python
import plotly.graph_objects as go

fig = go.Figure(data=[go.Scatter(
    x=male["Salary"], y=male['Experience_Years'],
    mode='markers',
    marker_size=[5,10,15,20])
])

fig.show()
```

![Alt text](newplot.png)

```python
species = ('Male','Female')
avgs = {
    'Below-Average': np.array([male['Experience_Years'].mean(), female['Experience_Years'].mean()]),
    'Above-Average': np.array([male['Experience_Years'].max(), male['Experience_Years'].max()]),
}
width = 0.6  # the width of the bars: can also be len(x) sequence


fig, ax = plt.subplots()
bottom = np.zeros(2)

for avg, avgs in avgs.items():
    p = ax.bar(species, avgs, width, label=avg, bottom=bottom)
    bottom += avgs

    ax.bar_label(p, label_type='center')

ax.set_title('Number of penguins by sex')
ax.legend()

plt.show()
```
