```python
from tkinter import *
from tkinter import font
from tkinter import messagebox
regform=Tk()
regform.title="Registration Form"
regform.geometry('800x1100')

heading=Label(text="Registration Form",foreground="black",width=40,font=font.Font(size=18))
heading.place(x=30,y=30)

lb1=Label(regform,text="Enter Name",foreground="black",width=10,font=("arial",12))
lb1.place(x=20,y=120)
name=StringVar()
en1=Entry(regform)
en1.place(x=200,y=120)

lb2=Label(regform,text="Enter Email",foreground="black",width=10,font=("arial",12))
lb2.place(x=20,y=160)
email=StringVar()
en2=Entry(regform)
en2.place(x=200,y=160)

lb3=Label(regform,text="Contact Number",foreground="black",width=12,font=("arial",12))
lb3.place(x=20,y=200)
contact=StringVar()
en3=Entry(regform)
en3.place(x=200,y=200)

lb4=Label(regform,text="Select Gender",foreground="black",width=12,font=("arial",12))
lb4.place(x=20,y=240)
var=StringVar()
var.set("MALE")
Radiobutton(regform, text="Male",padx=5,variable=var,value="MALE").place(x=200,y=240)
Radiobutton(regform, text="Female",padx=5,variable=var,value="FEMALE").place(x=280,y=240)
Radiobutton(regform, text="Other",padx=5,variable=var,value="OTHERS").place(x=360,y=240)

lb5=Label(regform,text="Select Country",foreground="black",width=12,font=("arial",12))
lb5.place(x=20,y=280)
list_of_country=("USA","India","Russia","Japan","Australia")
cv=StringVar()
drplist=OptionMenu(regform,cv,*list_of_country)
drplist.config(width=15)
cv.set("USA")
drplist.place(x=200,y=280)

lb6=Label(regform,text="Enter Password",foreground="black",width=12,font=("arial",12))
lb6.place(x=20,y=320)
password=StringVar()
en6=Entry(regform,show="*")
en6.place(x=200,y=320)

lb7=Label(regform,text="Re-Enter Password",foreground="black",width=15,font=("arial",12))
lb7.place(x=20,y=360)
cpassword=StringVar()
en7=Entry(regform,show="*")
en7.place(x=200,y=360)

cb1=StringVar()
cb2=StringVar()
cb3=StringVar()
cb4=StringVar()
english=Checkbutton(regform,text="English", variable=cb1).place(x=20,y=400)
Kannada=Checkbutton(regform,text="Kannada", variable=cb2).place(x=100,y=400)
Telugu=Checkbutton(regform,text="Telugu", variable=cb3).place(x=200,y=400)
Tamil=Checkbutton(regform,text="Tamil", variable=cb4).place(x=300,y=400)

Button(regform,text="Register",width=10).place(x=200,y=450)

regform.mainloop()
messagebox.showinfo("ShowInfo","Registration form is created Successfully..")
print(var.get())
print(cv.get())
print(cb1.get())
print(cb2.get())
print(cb3.get())
print(cb4.get())
```
