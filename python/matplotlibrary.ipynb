```python
setosa=iris[iris['Species']=='Iris-setosa']
versicolor=iris[iris['Species']=='Iris-versicolor']
virginica=iris[iris['Species']=='Iris-virginica']
print(setosa)
# print(versicolor)
# print(virginica)
```

```python
setosa["SepalWidthCm"].plot(kind="hist", label="Setosa", color="r", fontsize="10", density=False)
plt.title("Scatter Plot - Histogram")
plt.xlabel("SepalWidthCm")
plt.show()
```

```python
setosa.describe()
```

```python
#scatter Plot
setosa.plot(x="SepalLengthCm",y="PetalLengthCm", kind="scatter", label="Setosa", color="b", fontsize="10")
plt.title("Scatter Plot - Setosa")
plt.axis(0,50,2,5)
plt.xlabel("SepalLengthCm")
plt.ylabel("PetalLengthCm")
plt.show()
```

```python
#line plot
setosa['SepalWidthCm'].plot(kind='line',label="setosa",color="black",linewidth='1.5',ls="--")
plt.title("Line Plot - Setosa")
plt.xlabel("SepalLengthCm")
plt.show()
```

```python
setosa['SepalWidthCm'].plot(kind='line',label="setosa",color="black",linewidth='1.5',ls="--")
versicolor['SepalWidthCm'].plot(kind='line',label="versicolor",color="red",linewidth='1.5',ls="-")
virginica['SepalWidthCm'].plot(kind='line',label="virginica",color="blue",linewidth='1.5',ls=":")
plt.title("Sepecies wise - Line Plot")
plt.xlabel("SepalWidthCm")
plt.show()
```

```python
Type=list(iris['Species'].unique())
count=list(iris['Species'].value_counts())
Avg=list(iris.groupby('Species')["SepalLengthCm"].mean())
plt.bar(Type, Avg, color=['Maroon','yellow','blue'],width=0.4)
plt.title('Bar - Plot')
plt.xlabel('SepalLengthCm')
plt.ylabel('AverageSepalLengthCm')
plt.show()
```

```python
Type=list(iris['Species'].unique())
count=list(iris['Species'].value_counts())
Avg=list(iris.groupby('Species')["SepalLengthCm"].mean())
plt.barh(Type, Avg, color=['Maroon','yellow','blue'],height=0.4)
plt.title('Bar - Plot')
plt.xlabel('SepalLengthCm')
plt.ylabel('AverageSepalLengthCm')
plt.show()
```

```python
e=(0,0.1,0)
Avg=list(iris.groupby('Species')['SepalLengthCm'].mean())
plt.pie(Avg,explode=e,labels=['setosa','versicolor','virginica'])
plt.title('Pie Chart')
plt.legend(title="Iris Flowers:")
plt.legend(loc="upper right")
plt.show()
```
